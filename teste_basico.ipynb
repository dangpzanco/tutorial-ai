```python
# Usando funções
num_samples = 21
x = np.linspace(0,1, num_samples)
print(x)
```

```python
y = x ** 2
print(y)
```

```python
# Plotando
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("$y = x^2$")
plt.show()
```

```python
#Calculando a exponencial (e^)
y = np.exp(x)
print(y)
```

```python
# Plotando
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("$y = e^x$")
plt.show()
```

```python
y1 = x ** 2
y2 = np.exp(x)

# Plotando múltiplas linhas
plt.figure(figsize=(10,7))

plt.plot(x,y1)
plt.plot(x,y2)

plt.xlabel('x')
plt.ylabel('y')
plt.title("$f(x)$")
plt.legend(["$y = x^2$", "$y = e^x$"])
plt.show()
```

```python
# Carregando dados de um dataset
df = pd.read_csv('diabetes.csv')
df.head(5)
```

```python
# Calculando a média
age = df['Age'].values
mean_age = age.mean()
mean_age = df['Age'].mean() # equivalente

print(f'Idade Média: {mean_age:0.2f} anos')
```
